# Splitting the Normal Transaction and Fraud Transaction

```python
legit = data[data['Class'] == 0]
fraud = data[data['Class'] == 1]
```

```python
# Convert all columns except 'Class' to numeric, handling errors
for col in data.columns:
    if col != 'Class':
        try:
            data[col] = pd.to_numeric(data[col], errors='coerce')
        except ValueError:
            print(f"Warning: Could not convert column '{col}' to numeric.")

# Now, recalculate the group means
data.groupby('Class').mean()
```


<span style="font-size:20px">Since our data set is very unbalanced so we are taking sample of Normal Transaction to make our data set balanced</span>

```python
legit_sample = legit.sample(n=492)
legit_sample
```

<span style="font-size:20px">Concatenating Sample data set and fraud data set as it is now balanced data set </span>

```python
concat_data = pd.concat([legit_sample , fraud],axis = 0)
concat_data
```

```python
 concat_data['Class'].value_counts()
```

```python
# Convert all columns except 'Class' to numeric, handling errors
for col in concat_data.columns:
    if col != 'Class':
        try:
            concat_data[col] = pd.to_numeric(concat_data[col], errors='coerce')
        except ValueError:
            print(f"Warning: Could not convert column '{col}' to numeric.")

# Now, recalculate the group means
concat_data.groupby('Class').mean()
```

```python
X = concat_data.drop(columns='Class' , axis =1)
Y = concat_data['Class']
```

<span style="font-size:20px">X-----> Contains all the data columns except Class column</span>

<span style="font-size:20px">Y-----> Contains Labels 0 and 1</span>

```python
X
```

```python
Y
```

<span style="font-size:20px">Training the model</span>

```python
X_train ,X_test, Y_train, Y_test = train_test_split(X,Y,test_size =0.2,stratify=Y,random_state=2)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train , Y_train)
```

<span style="font-size:20px">Checking the accuracy of training data</span>

```python
X_train_prediction = model.predict(X_train)
Accuracy_Train = accuracy_score(X_train_prediction,Y_train)
Accuracy_Train
```

<span style="font-size:20px">Checking the accuracy of test data</span>

```python
X_test_prediction = model.predict(X_test)
Accuracy_Test = accuracy_score(X_test_prediction , Y_test)
Accuracy_Test
```

# TEST INTERFACE

```python
print("\n--- Credit Card Fraud Detection Test Interface ---")
print("Enter transaction details below. Type 'quit' at any prompt to exit.")

# Get the feature names in the order the model expects them
feature_names = X.columns.tolist()

while True:
    input_data_list = []
    try:
        print("\nPlease enter the following transaction features:")
        for feature in feature_names:
            user_input = input(f"  Enter value for {feature}: ")
            if user_input.lower() == 'quit':
                raise KeyboardInterrupt # Use an exception to break out cleanly

            # Attempt to convert input to float
            input_data_list.append(float(user_input))

        # Convert list to NumPy array and reshape for prediction (1 row, 30 columns)
        input_data_array = np.array(input_data_list).reshape(1, -1)

        # Make prediction
        prediction = model.predict(input_data_array)
        prediction_proba = model.predict_proba(input_data_array) # Get probabilities

        # Display result
        print("\n--- Prediction Result ---")
        if prediction[0] == 0:
            print("  Prediction: Legitimate Transaction")
        else:
            print("  Prediction: >> Fraudulent Transaction <<")

        print(f"  Probability -> [Legitimate: {prediction_proba[0][0]:.4f}, Fraudulent: {prediction_proba[0][1]:.4f}]")
        print("-------------------------\n")

    except ValueError:
        # Handle cases where input is not a valid number
        print("\nError: Invalid input. Please enter numeric values only.")
        print("Restarting transaction input...\n")
        continue # Skip rest of the loop and start asking for input again

    except KeyboardInterrupt:
        # Handle user typing 'quit'
        print("\nExiting test interface.")
        break # Exit the while loop

    except Exception as e:
        # Catch any other unexpected errors
        print(f"\nAn unexpected error occurred: {e}")
        print("Exiting test interface.")
        break

print("--- Interface Closed ---")
```
