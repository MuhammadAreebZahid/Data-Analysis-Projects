```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/tested.csv")
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna("No age", inplace = True)
df
df.dropna(inplace=True)
df
```

```python
df.drop_duplicates()
```

```python
def remove_outliers(df, col):
    if pd.api.types.is_numeric_dtype(df[col]):
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        # Return filtered dataframe
        return df[(df[col] >= lower) & (df[col] <= upper)]
    else:
        print(f"âš ï¸ Column '{col}' is not numeric.")
        return df

# Apply only on numeric columns
df = remove_outliers(df, 'Fare')
df = remove_outliers(df, 'Age')
```

```python
sns.countplot(x='Sex', data=df, color='green')
plt.title('Sex Distribution')
plt.show()
```

```python
sns.countplot(x='Pclass', data=df, color='red')
plt.title("Passenger Class Distribution")
plt.show()
```

```python
# It will Convert 'Age' column to numeric, replace non-numeric with NaN
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# This one Remove rows with NaN values in 'Age' if necessary
df.dropna(subset=['Age'], inplace=True)

plt.figure(figsize=(8, 5))
df['Age'].hist(bins=20, color='darkred',edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# It will Convert 'Fare' column to numeric, replace non-numeric with NaN
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')

# This one Remove rows with NaN values in 'Fare' if necessary
df.dropna(subset=['Fare'], inplace=True)

df['Fare'].hist(bins=20)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
